```python
from pathlib import Path

import pandas as pd
```

## ค้นหา root ของโครงการ

รองรับการเปิด notebook จาก `notebooks/` หรือจาก root ของโครงการ โดยตรวจหา `data/raw/` ที่มีอยู่จริง

```python
def find_project_root():
    current_path = Path.cwd().resolve()
    candidate_paths = [current_path]
    candidate_paths.extend(current_path.parents)

    for candidate_path in candidate_paths:
        if (candidate_path / "data" / "raw").exists():
            return candidate_path

    raise FileNotFoundError("ไม่พบ root ของโครงการที่มี data/raw")


PROJECT_ROOT = find_project_root()
```

## กำหนด path หลัก

cell นี้กำหนดตำแหน่งจัดเก็บข้อมูลดิบจาก root ที่ตรวจพบ

## ตรวจไฟล์ข้อมูลดิบตาม registry

cell นี้ตรวจว่าชื่อไฟล์ที่บันทึกไว้มีอยู่จริงในโฟลเดอร์ `data/raw/` หรือไม่

```python
local_file_status_records = []

for source_record in source_registry:
    local_path = PROJECT_ROOT / source_record["local_file"]
    local_file_status_records.append(
        {
            "dataset_id": source_record["dataset_id"],
            "local_file": source_record["local_file"],
            "file_exists": local_path.is_file(),
            "file_size_bytes": local_path.stat().st_size if local_path.is_file() else None,
        }
    )
```

## แสดงสถานะไฟล์ข้อมูลดิบ

```python
local_file_status_df = pd.DataFrame(local_file_status_records)
local_file_status_df
```
